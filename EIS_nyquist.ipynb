```python
import pandas as pd
import numpy as np

import plotly.graph_objects as go
import plotly.express as px
from plotly.subplots import make_subplots
```

```python
df_zyklen = {
    "0" : pd.read_csv("0_zyklen.csv", header = 1),
    "50" : pd.read_csv("50_zyklen.csv", header = 1),
    "100" : pd.read_csv("100_zyklen.csv", header = 1),
    "150" : pd.read_csv("150_zyklen.csv", header = 1),
    "200" : pd.read_csv("200_zyklen.csv", header = 1),
    "250" : pd.read_csv("250_zyklen.csv", header = 1),
    "300" : pd.read_csv("300_zyklen.csv", header = 1),
    "350" : pd.read_csv("350_zyklen.csv", header = 1),
    "400" : pd.read_csv("400_zyklen.csv", header = 1),
}

for k, v in enumerate(df_zyklen):
    df_zyklen["{}".format(v)]["Im/Ohm"] = df_zyklen["{}".format(v)]["Im/Ohm"]*(-1)
df_zyklen["50"]
```

```python
traces = [0]*9

for k, v in enumerate(df_zyklen):
    #print(k)
    traces[k] = go.Scatter(x = df_zyklen["{}".format(v)]["Re/Ohm"], y = df_zyklen["{}".format(v)]["Im/Ohm"],mode = "lines")


fig = make_subplots(rows = 1, cols = 1)
for k, v in enumerate(df_zyklen):
    fig.add_trace(traces[k])
    
fig.show()
```
